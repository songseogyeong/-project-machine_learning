# Heart Attack : 심장 마비

## 환자의 심장 마비 예측
#### ○ 이진 분류 (Binary Classification)
- 병원에서 수집된 환자의 심장 마비 발생 여부 데이터 세트
- 환자의 상태 데이터를 제공하고 심장마비 또는 없음으로 분류된다.

#### ○ Feature (8개)
- Age: 나이
- Gender: 성별
  ※ 남성: 1, 여성: 0
- Heart rate: 심박수
- Systolic blood pressure: 수축기 혈압
- Diastolic blood pressure: 이완기 혈압
- Blood sugar: 혈당
- CK-MB: 심장 효소  
  ※ 정상수치: 0.4 ng/ml 이하
- Troponin: 트로포닌  
  ※ 정상범위: 0~3mg/dl 이하

#### ○ target
- Result: 심장마비 여부  
  ※ negative: 정상환자(음성), positive: 심장마비(양성)

#### 📊 환자 성비

```python
# 전체 인원 성비
gender_counts = ha_df['Gender'].value_counts().reset_index()
gender_counts['Gender'] = gender_counts['Gender'].map({1: 'male', 0: 'female'})
# 정상 환자 성비
gender_ne_counts = ha_df[ha_df['Result'] == 'negative']['Gender'].value_counts().reset_index()
gender_ne_counts['Gender'] = gender_ne_counts['Gender'].map({1: 'male', 0: 'female'})
# 심장 마비 환자 성비
gender_po_counts = ha_df[ha_df['Result'] == 'positive']['Gender'].value_counts().reset_index()
gender_po_counts['Gender'] = gender_po_counts['Gender'].map({1: 'male', 0: 'female'})
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 10))

# plt.rcParams['font.size'] = 12

# 부채꼴 스타일
wedgeprops = {'width': 0.7, 'edgecolor': 'w', 'linewidth': 1}

# 컬러맵 가져오기
cmap = plt.get_cmap('Pastel1')
# 컬러맵의 색상 선택
all_colors = cmap(np.array([0,4]))
outer_colors = cmap(np.array([1,2]))

# 그래프 데이터 가져오기
counts = gender_counts['count']
labels = gender_counts['Gender']
ne_counts = gender_ne_counts['count']
ne_labels = gender_ne_counts['Gender']
po_counts = gender_po_counts['count']
po_labels = gender_po_counts['Gender']

# 그래프 그리기
axes[0].pie(counts, labels=labels, colors=all_colors, autopct='%1.1f%%', startangle=90, wedgeprops=wedgeprops)
axes[0].set_title('전체 환자 비율')
axes[1].pie(ne_counts, labels=ne_labels, colors=outer_colors, autopct='%1.1f%%', startangle=90, wedgeprops=wedgeprops)
axes[1].set_title('정상 환자 비율')
axes[2].pie(po_counts, labels=po_labels, colors=outer_colors, autopct='%1.1f%%', startangle=90, wedgeprops=wedgeprops)
axes[2].set_title('심장마비 환자 비율')

display(pd.concat([gender_counts.reset_index(drop=True), gender_ne_counts.reset_index(drop=True), gender_po_counts.reset_index(drop=True)], axis=1))

plt.tight_layout()
plt.show()
```

#### 📊 0을 가지는 데이터 확인

```python
# 피처 값이 0인 데이터의 개수와 비율 확인

# 피처 zero_feaures에 담기
zero_features = ['Heart rate', 'CK-MB', 'Troponin']

# 데이터 프레임의 전체 행의 수를 가져와 담기
total_count = ha_df.shape[0]

# 각 zero_features들을 반복하여 feature에 하나씩 담기:
for feature in zero_features:
    # 각 피처에서  값이 0이 되는 행을 가져오고 행의 개수를 세어 변수에 담기
    zero_count = ha_df[ha_df[feature] == 0][feature].count()
    # 각 피처명, 값이 0인 데이터의 개수, 전체 비율 출력
    print(f'{feature}: {zero_count}건, {np.round(zero_count / total_count * 100, 4)}%')
```

#### 💡 분석
○ 심박수
- 전체 환자의 심박수 분포는 0~200bpm 사이로 고르게 나오나, 1000bpm이 넘어가는 수치가 나타남.

<br>

○ 수축기 혈압
- 수축기 혈압의 분포는 고르게 나타남.

<br>

○ 이완기 혈압
- 이완기 혈압의 푼보는 전반적으로 고르게 나타나지만, 일부 수치가 높게 나타남.

<br>

○ 혈당
- 혈당은 100mg/dl에서 많은 분포가 나타나며, 200mg/dl이 넘는 수치도 나타남.

<br>

○ CK-MB (크레아틴 키나제-MB)
- 크레아틴 키나제(심장 근육 손상을 나타내는 효소)는 0~10ml/dl에서 많이 나타남.

<br>

○ Troponin (트로포닌)
- 트로포닌(심근 손상 여부를 사용하는데 판단되는 단백질)는 0~2ng/ml에서 많이 나타남.

#### 📌 시사점
○ 심박수
- 보통 사람의 최대 심박수는 200bpm으로 1000bpm이 넘어가는 수치는 정확한 수치가 아니라 판단되므로 이상치 제거가 필요함.

<br>

○ 수축기 혈압
- 대게 사람의 최고혈압은 170mmHg, 최저 혈압은 55mmHg로 나타나며, 큰 이상치는 보이지 않음.
- 50mmHg미만으로 떨어지는 수치는 환자 상태에 따라 이상치 판단 여부가 달라질 것으로 판단되며,  
  정상 환자라면 50mmHg는 측정되기 어려운 수치로 이상치로 제거해야 함.
- 200mmHg 넘는 환자는 고혈압 환자 수치로 판단되어 이상치로 판단하지 않음.

<br>

○ 이완기 혈압
- 이완기 혈압 100mmHg 이상은 고혈압 환자의 수치로 이상치로 판단하지 않지만, 120mmHg이 초과되는 수치는 정상 수치가 아니라고 판단되며, 제거가 필요함.

<br>

○ 혈당
- 혈당 수치 200mg/dl가 넘는 환자는 당뇨 환자로 별도의 이상치를 제거하지 않음.

<br>

○ CK-MB(크레아틴 키나제-MB)
- 심장 근육의 손상 정상치는 0ml/dl ~ 6.73ml/dl이지만, 심장질환으로 인해 정상치를 넘어가는 데이터가 있을 것이라 판단됨.
- 100ml/dl 이상부터는 비정상적인 수치로 판단되며, 이상치로 제거가 필요함.

<br>

○ Troponin (트로포닌)
- 트로포닌의 정상수치는 0ng/ml ~ 0.4ng/ml 이하지만, 심장질환으로 인해 정상치를 넘어가는 데이터가 있을 것이라 판단되므로 별도의 이상치 제거를 하지 않음.

```python
import numpy as np

# Classifier의 Decision Boundary를 시각화 하는 함수
def visualize_boundary(model, X, y):
    fig,ax = plt.subplots()
    
    # 학습 데이타 scatter plot으로 나타내기
    ax.scatter(X.iloc[:, 0], X.iloc[:, 1], c=y, s=25, cmap='rainbow', edgecolor='k',
               clim=(y.min(), y.max()), zorder=3)
    ax.axis('tight')
    ax.axis('off')
    xlim_start , xlim_end = ax.get_xlim()
    ylim_start , ylim_end = ax.get_ylim()
    
    # 호출 파라미터로 들어온 training 데이타로 model 학습 . 
    model.fit(X.values, y)
    # meshgrid 형태인 모든 좌표값으로 예측 수행. 
    xx, yy = np.meshgrid(np.linspace(xlim_start,xlim_end, num=200),np.linspace(ylim_start,ylim_end, num=200))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    
    # contourf() 를 이용하여 class boundary 를 visualization 수행. 
    n_classes = len(np.unique(y))
    contours = ax.contourf(xx, yy, Z, alpha=0.3,
                           levels=np.arange(n_classes + 1) - 0.5,
                           cmap='rainbow',
                           zorder=1)
```

```python
features = ha_sta_df[['CK-MB', 'Troponin']] 
dtc_db = DecisionTreeClassifier(random_state=124)
visualize_boundary(dtc_db, features, targets)
```

- 정확도만으로는 모델의 성능을 완전히 평가할 수 없으므로 오차행렬 지표를 통해 성능 평가를 진행

#### 📊 오차행렬

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, ConfusionMatrixDisplay

# 타겟 데이터와 예측 객체를 전달받는다.
def get_evaluation_by_custom_proba(y_test, prediction, visualize=False):
#     오차 행렬
    confusion = confusion_matrix(y_test, prediction)
#     정확도
    accuracy = accuracy_score(y_test , prediction)
#     정밀도
    precision = precision_score(y_test , prediction)
#     재현율
    recall = recall_score(y_test , prediction)
#     F1 score
    f1 = f1_score(y_test, prediction)
#     ROC-AUC
    roc_auc = roc_auc_score(y_test, prediction)
    
    print('오차 행렬')
    print(confusion)
    print('정확도: {0:.4f}, 정밀도: {1:.4f}, 재현율: {2:.4f}, F1: {3:.4f}, ROC-AUC: {4:.4f}'.format(accuracy, precision, recall, f1, roc_auc))
    print("#" * 80)
    
    if visualize:
        fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8,4))
        titles_options = [("Confusion matrix", None), ("Normalized confusion matrix", "true")]

        for (title, normalize), ax in zip(titles_options, axes.flatten()):
            disp = ConfusionMatrixDisplay.from_predictions(y_test, prediction, ax=ax, cmap=plt.cm.Blues, normalize=normalize)
            disp.ax_.set_title(title)
        plt.show()
```

```python
get_evaluation_by_custom_proba(y_test, prediction, visualize=True)
```

### Ⅹ. 정밀도, 재현률 수립

#### 📊 심장마비 진단
- 양성(Positive): 심장마비 환자를 의미
- 음성(Negative): 정상 환자를 의미

<br>

- True Positive (TP): 모델이 심장마비 환자를 심장마비로 올바르게 분류한 경우.  
- False Positive (FP): 모델이 정상 환자를 심장마비로 잘못 분류한 경우(오진, error I).  
- True Negative (TN): 모델이 정상 환자를 정상으로 올바르게 분류한 경우.  
- False Negative (FN): 모심장마비 고암 환자를 정상으로 잘못 분류한 경우(놓침, erro).  

<br>

- 정밀도 (Precision): 모델이 심장마비로 예측한 환자 중에서 실제로 심장마비인 환자의 비율을 나타낸다.  
  즉, 양성 클래스로 예측한 결과 중 얼마나 많은 것이 진짜 양성인지 측정한다.
- 재현율 (Recall): 실제심장마비암 환자 중에서 모델이심장마비로로 정확하게 예측한 환자의 비율을 나타낸다.  
  즉, 모델이 얼마나 많은심장마비암 환자를 "놓치지 않고" 감지했는지 측정한다

<br>

임계치를 낮춰서 재현율을 높여주는 것이 적합하다.

#### 📊 F1 Score

```python
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

def precision_recall_curve_plot(y_test , proba_class1):

    precisions, recalls, thresholds = precision_recall_curve(y_test, proba_class1)
    
    # X축: threshold
    # Y축: 정밀도, 재현율 
    # 정밀도는 점선으로 표시
    plt.figure(figsize=(8,6))
    threshold_boundary = thresholds.shape[0]
    plt.plot(thresholds, precisions[0:threshold_boundary], linestyle='--', label='precision')
    plt.plot(thresholds, recalls[0:threshold_boundary],label='recall')
    
    # X축(threshold)의 Scale을 0 ~ 1 단위로 변경
    start, end = plt.xlim()
    plt.xticks(np.round(np.arange(start, end, 0.1),2))
    
    plt.xlabel('Threshold value'); plt.ylabel('Precision and Recall value')
    plt.legend()
    plt.grid()
    plt.show()
    
precision_recall_curve_plot(y_test, dtc.predict_proba(X_test)[:, 1].reshape(-1, 1))
```

#### 📊 ROC Curve, AUC

```python
from sklearn.metrics import roc_curve

def roc_curve_plot(y_test , proba_class1):
#     임계값에 따른 FPR, TPR 값
    fprs, tprs, thresholds = roc_curve(y_test, proba_class1)

#     ROC Curve를 plot 곡선으로 그림. 
    plt.plot(fprs , tprs, label='ROC')
#     가운데 대각선 직선을 그림. 
#     TPR과 FPR이 동일한 비율로 떨어진다는 것은 모델이 양성과 음성을 구별하지 못한다는 것을 의미한다.
#     다른 분류기를 판단하기 위한 기준선으로 사용되며, 
#     대각선에 가까울 수록 예측에 사용하기 힘든 모델이라는 뜻이다.
    plt.plot([0, 1], [0, 1], 'k--', label='Standard')
    
    # X축(FPR)의 Scale을 0.1 단위로 변경
    start, end = plt.xlim()
    plt.xticks(np.round(np.arange(start, end, 0.1),2))
    plt.xlim(0,1); plt.ylim(0,1)
    plt.xlabel('FPR( 1 - Sensitivity )'); plt.ylabel('TPR( Recall )')
    plt.legend()
    plt.show()
    
roc_curve_plot(y_test, dtc.predict_proba(X_test)[:, 1].reshape(-1, 1))
```
