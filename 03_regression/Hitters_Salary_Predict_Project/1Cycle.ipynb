1986년과 1987년 시즌의 메이저 리그 야구 데이터.

<br>

#### features (19개)
- AtBat: 1986년 타석수
- Hits: 1986년 안타 수
- HmRun: 1986년 홈런 수
- Runs: 1986년 런수
- RBI : 1986년 타점
- Walks: 1986년 볼넷 수
- Years : 메이저리그 연식
- CAtBat : 선수 생활 중 타석 수
- CHits : 선수 생활 중 안타 수
- CHmRun : 선수 생활 중 홈런 수
- CRuns: 선수 생활 중 실점
- CRBI: 선수 생활 동안의 타점 수
- CWalks: 그의 선수 생활 동안의 볼넷 수
- League: 1986년 말 선수의 리그를 나타내는 A, N 레벨을 갖는 인자
- Division: 1986년 말 선수의 디비전을 나타내는 레벨 E, W를 가지는 인자
- PutOuts: 1986년 Puts Outs 수
- Assists: 1986년 어시스트 수
- Errors: 1986년 오류 수
- NewLeague: 1987년 초 선수의 리그를 나타내는 A, N 레벨을 갖는 인자

<br>

#### targets
- Salary: 개업일 1987년 연봉 수천 달러

#### 📊 데이터 전처리

```python
# 타겟 데이터 이상치 제거
from sklearn.preprocessing import StandardScaler

# 표준화 객체 생성
std = StandardScaler()
# 데이터 표준화 젼환
result = std.fit_transform(h_df[['Salary']])
# 표준화 데이터 프레임 생성
std_h_df = pd.DataFrame(result, columns=['Salary'])

# 타겟 데이터 이상치 제거
std_h_df = std_h_df[std_h_df.Salary.between(-1.96, 1.96)]

# 원본 데이터 프레임에도 타겟 데이터 이상치 제거 적용 후 새로운 데이터 프레임에 담기
pre_h_df = h_df.iloc[std_h_df.index].reset_index(drop=True)

pre_h_df
```

```python
# 데이터 분포 확인
pre_h_df.hist(figsize=(12, 15), color=cmap(np.array([1])), edgecolor='black')
```

```python
# 타겟 컬럼 분포 변환
from sklearn.preprocessing import PowerTransformer

ptf = PowerTransformer(standardize=False)
pre_h_df['Salary'] = ptf.fit_transform(pre_h_df[['Salary']])
pre_h_df
```

```python
# 타겟 컬럼 분포 변환 확인
pre_h_df.Salary.hist(color=cmap(np.array([1])), edgecolor='black')
```

```python
from sklearn.preprocessing import LabelEncoder

# 데이터 프레임 복제
enc_h_df = pre_h_df.copy()
encoders = {}

# 문자열 컬럼 추출
columns = ['League', 'Division', 'NewLeague']

# 반복하여 컬럼 인코딩 처리:
for column in columns:
    # 레이블 인코더 객체 생성
    encoder = LabelEncoder()
    # 문자열 데이터 정수로 형변환
    result = encoder.fit_transform(enc_h_df[column])
    # 형변환 값으로 대체
    enc_h_df[column] = result
    # 원본 데이터 담기
    encoders[column] = encoder.classes_

# 원본 데이터 확인
print(encoders)
```

```python
# 타겟 컬럼 위치 변경을 위해 새로운 컬럼에 데이터 담기
enc_h_df['target'] = enc_h_df.Salary
# 기존 컬럼은 제거
enc_h_df = enc_h_df.drop(labels='Salary', axis=1)
enc_h_df
```

```python
# 상관관계 확인
enc_h_df.corr()['target'].sort_values(ascending=False)[1:]
```

```python
# 선수의 경력 데이터들의 평균 값으로 대체
enc_h_df['OrtCAtBat'] = enc_h_df['CAtBat'] / enc_h_df['Years']
enc_h_df['OrtCHits'] = enc_h_df['CHits'] / enc_h_df['Years']
enc_h_df['OrtCHmRun'] = enc_h_df['CHmRun'] / enc_h_df['Years']
enc_h_df['OrtCruns'] = enc_h_df['CRuns'] / enc_h_df['Years']
enc_h_df['OrtCRBI'] = enc_h_df['CRBI'] / enc_h_df['Years']
enc_h_df['OrtCWalks'] = enc_h_df['CWalks'] / enc_h_df['Years']
```

```python
# 불필요한 피처, 사용이 완료된 피처 및 상관관계가 낮은 제거
columns = ['Errors', 'NewLeague', 'League', 'Years']

enc_h_df = enc_h_df.drop(labels=columns, axis=1)
enc_h_df
```

#### 📊 회귀

```python
import numpy as np
from sklearn.metrics import mean_squared_error, mean_squared_log_error, r2_score

def get_evaluation(y_test, prediction):
    MSE = mean_squared_error(y_test, prediction)
    RMSE = np.sqrt(MSE)
    MSLE = mean_squared_log_error(y_test, prediction)
    RMSLE = np.sqrt(MSLE)
    R2 = r2_score(y_test, prediction)
    print('MSE: {:.4f}, RMSE: {:.4f}, MSLE: {:.4f}, RMSLE: {:.4f}, R2: {:.4f}'\
         .format(MSE, RMSE, MSLE, RMSLE, R2))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# 데이터 세트 분리
# 피처, 타겟 데이터 분리
features, targets = enc_h_df.iloc[:, :-1], enc_h_df.iloc[:, -1]

# 학습/테스트 및 문제/정답 데이터 세트 분리
X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

# 선형 회귀 모델
l_r = LinearRegression()
# 훈련
l_r.fit(X_train, y_train)
```

```python
# 예측
prediction = l_r.predict(X_test)
# 평가
get_evaluation(y_test, prediction)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error, mean_squared_log_error, r2_score

def get_evaluation(y_test, prediction):
    MSE = mean_squared_error(y_test, prediction)
    RMSE = np.sqrt(MSE)
    R2 = r2_score(y_test, prediction)
    print('MSE: {:.4f}, RMSE: {:.4f}, R2: {:.4f}'\
         .format(MSE, RMSE, R2))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures

# 데이터 세트 분리
# 피처, 타겟 데이터 분리
features, targets = enc_h_df.iloc[:, :-1], enc_h_df.iloc[:, -1]

# 차원 확장 (1차원)
poly_features = PolynomialFeatures(degree=2).fit_transform(features)

# 학습/테스트 및 문제/정답 데이터 세트 분리
X_train, X_test, y_train, y_test = \
train_test_split(poly_features, targets, test_size=0.2, random_state=124)

# 선형 회귀 모델
l_r = LinearRegression()
# 훈련
l_r.fit(X_train, y_train)
```

```python
# 예측
prediction = l_r.predict(X_test)
# 평가
get_evaluation(y_test, prediction)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.model_selection import train_test_split

# 데이터 세트 분리
# 피처, 타겟 데이터 분리
features, targets = enc_h_df.iloc[:, :-1], enc_h_df.iloc[:, -1]

# 학습/테스트 및 문제/정답 데이터 세트 분리
X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

# 회귀 모델 담기
dt_r = DecisionTreeRegressor(random_state=124)
rf_r = RandomForestRegressor(random_state=124, n_estimators=1000)
gb_r = GradientBoostingRegressor(random_state=124)
xgb_r = XGBRegressor(random_state=124)
lgb_r = LGBMRegressor(random_state=124)

models = [dt_r, rf_r, gb_r, xgb_r, lgb_r]

# 모델 별 학습 진행:
for model in models:
    # 모델 학습
    model.fit(X_train, y_train)
    # 예측
    prediction = model.predict(X_test)
    # 모델명 출력 (모델명과 평가 결과가 같이 출력되도록)
    print(model.__class__.__name__)
    # 평가
    get_evaluation(y_test, prediction)
```
