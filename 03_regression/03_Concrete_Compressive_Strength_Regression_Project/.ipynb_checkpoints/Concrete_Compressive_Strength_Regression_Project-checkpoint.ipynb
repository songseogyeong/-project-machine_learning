### üåü 1Cycle

#### üìä ÏÑ†Ìòï ÌöåÍ∑Ä

```python
import numpy as np
from sklearn.metrics import mean_squared_error, mean_squared_log_error, r2_score

def get_evaluation(y_test, prediction):
    MSE = mean_squared_error(y_test, prediction)
    RMSE = np.sqrt(MSE)
    MSLE = mean_squared_log_error(y_test, prediction)
    RMSLE = np.sqrt(MSLE)
    R2 = r2_score(y_test, prediction)
    print('MSE: {:.4f}, RMSE: {:.4f}, MSLE: {:.4f}, RMSLE: {:.4f}, R2: {:.4f}'\
         .format(MSE, RMSE, MSLE, RMSLE, R2))
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ Î∂ÑÎ¶¨
# ÌîºÏ≤ò, ÌÉÄÍ≤ü Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨
features, targets = pre_c_df.iloc[:, :-1], pre_c_df.iloc[:, -1]

# ÌïôÏäµ/ÌÖåÏä§Ìä∏ Î∞è Î¨∏Ï†ú/Ï†ïÎãµ Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ Î∂ÑÎ¶¨
X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

# ÏÑ†Ìòï ÌöåÍ∑Ä Î™®Îç∏
l_r = LinearRegression()
# ÌõàÎ†®
l_r.fit(X_train, y_train)
```

```python
# ÏòàÏ∏°
prediction = l_r.predict(X_test)
# ÌèâÍ∞Ä
get_evaluation(y_test, prediction)
```

#### üìä Îã§Ï§ë ÌöåÍ∑Ä

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures

# Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ Î∂ÑÎ¶¨
# ÌîºÏ≤ò, ÌÉÄÍ≤ü Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨
features, targets = pre_c_df.iloc[:, :-1], pre_c_df.iloc[:, -1]

# Ï∞®Ïàò ÌôïÏû•
poly_features = PolynomialFeatures(degree=3).fit_transform(features)

# ÌïôÏäµ/ÌÖåÏä§Ìä∏ Î∞è Î¨∏Ï†ú/Ï†ïÎãµ Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ Î∂ÑÎ¶¨
X_train, X_test, y_train, y_test = \
train_test_split(poly_features, targets, test_size=0.2, random_state=124)

# ÏÑ†Ìòï ÌöåÍ∑Ä Î™®Îç∏
l_r = LinearRegression()
# ÌõàÎ†®
l_r.fit(X_train, y_train)
```

```python
# ÏòàÏ∏°
prediction = l_r.predict(X_test)
# ÌèâÍ∞Ä
get_evaluation(y_test, prediction)
```

####  üìä ÌöåÍ∑Ä Î™®Îç∏

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.model_selection import train_test_split

# Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ Î∂ÑÎ¶¨
# ÌîºÏ≤ò, ÌÉÄÍ≤ü Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨
features, targets = pre_c_df.iloc[:, :-1], pre_c_df.iloc[:, -1]

# ÌïôÏäµ/ÌÖåÏä§Ìä∏ Î∞è Î¨∏Ï†ú/Ï†ïÎãµ Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ Î∂ÑÎ¶¨
X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

# ÌöåÍ∑Ä Î™®Îç∏ Îã¥Í∏∞
dt_r = DecisionTreeRegressor(random_state=124)
rf_r = RandomForestRegressor(random_state=124, n_estimators=1000)
gb_r = GradientBoostingRegressor(random_state=124)
xgb_r = XGBRegressor(random_state=124)
lgb_r = LGBMRegressor(random_state=124)

models = [dt_r, rf_r, gb_r, xgb_r, lgb_r]

# Î™®Îç∏ Î≥Ñ ÌïôÏäµ ÏßÑÌñâ:
for model in models:
    # Î™®Îç∏ ÌïôÏäµ
    model.fit(X_train, y_train)
    # ÏòàÏ∏°
    prediction = model.predict(X_test)
    # Î™®Îç∏Î™Ö Ï∂úÎ†• (Î™®Îç∏Î™ÖÍ≥º ÌèâÍ∞Ä Í≤∞Í≥ºÍ∞Ä Í∞ôÏù¥ Ï∂úÎ†•ÎêòÎèÑÎ°ù)
    print(model.__class__.__name__)
    # ÌèâÍ∞Ä
    get_evaluation(y_test, prediction)
```

```python
from statsmodels.api import OLS

# Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ Î∂ÑÎ¶¨
# ÌîºÏ≤ò, ÌÉÄÍ≤ü Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨
features, targets = pre_c_df.iloc[:, :-1], pre_c_df.iloc[:, -1]

# OLS Í∞ùÏ≤¥ ÏÉùÏÑ±
model = OLS(targets, features)
# Î™®Îç∏ ÌõàÎ†® Î∞è Ï†ïÎ≥¥ Ï∂úÎ†•
print(model.fit().summary())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def get_vif(features):
    vif = pd.DataFrame()
    vif['vif_score'] = [variance_inflation_factor(features, i) for i in range(features.shape[1])]
    vif['features'] = features.columns
    return vif
```

```python
# VIF ÌôïÏù∏
get_vif(features)
```

### üåü 5Cycle
GridSearchCVÎ•º ÌÜµÌï¥ ÏµúÏ†ÅÏùò ÌïòÏù¥ÌçºÌååÎùºÎØ∏ÌÑ∞ Ï°∞Ìï©ÏùÑ Ï∞æÍ≥† ÏÑ±Îä•ÏùÑ Îçî Ìñ•ÏÉÅÏãúÌÇ¨ Ïàò ÏûàÎäîÏßÄ ÌôïÏù∏

```python
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold

# Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ Î∂ÑÎ¶¨
# ÌîºÏ≤ò, ÌÉÄÍ≤ü Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨
features, targets = df_m_df.iloc[:, :-1], df_m_df.iloc[:, -1]

# Î¨∏Ï†ú/Ï†ïÎãµ Î∞è ÌïôÏäµ/ÌõàÎ†® Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨
X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

# Í≤ÄÏ¶ù Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨
val_X_train, val_X_test, val_y_train, val_y_test = \
train_test_split(X_train, y_train, test_size=0.2, random_state=124)

# ÌöåÍ∑Ä Î™®Îç∏ Îã¥Í∏∞
k_xgb_r = XGBRegressor(random_state=124)

# ÌååÎùºÎØ∏ÌÑ∞ Í∞í Ï°∞Ï†ï
parameters = {'learning_rate': [0.1, 0.15, 0.2, 0.3], 'n_estimators': [150, 200, 250]}

# ÍµêÏ∞®Í≤ÄÏ¶ù
# n_splits: Îç∞Ïù¥ÌÑ∞Î•º Î™á Í∞úÏùò Ìè¥ÎìúÎ°ú ÎÇòÎàåÏßÄÎ•º Í≤∞Ï†ï (ÏùºÎ∞òÏ†ÅÏúºÎ°ú 5 ÎòêÎäî 10)
# shuffle: Î∂ÑÌï† Ï†Ñ Îç∞Ïù¥ÌÑ∞ ÌòºÌï© Ïó¨Î∂Ä 
kfold = KFold(n_splits=10, random_state=124, shuffle=True)

# ÌïôÏäµ Î∞è ÍµêÏ∞® Í≤ÄÏ¶ù Î™®Îç∏ ÏÑ§Ï†ï
grid_xgb_r = GridSearchCV(k_xgb_r, param_grid=parameters, cv=kfold, n_jobs=-1)

# ÌõàÎ†®
grid_xgb_r.fit(X_train, y_train)
```

```python
# ÌõàÎ†® Í≤∞Í≥º ÌôïÏù∏
result_df = pd.DataFrame(grid_xgb_r.cv_results_)[['params', 'mean_test_score', 'rank_test_score']]
display(result_df)
```

```python
# ÏµúÏ†ÅÏùò Î™®Îç∏ Îã¥Í∏∞
kf_xgb_r = grid_xgb_r.best_estimator_

# ÏòàÏ∏°
prediction = kf_xgb_r.predict(X_test)

# ÌèâÍ∞Ä
get_evaluation(y_test, prediction)
```

```python
# Ìé∏Ìñ• Ï∂îÍ∞ÄÎ•º ÏúÑÌï¥ Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑ Î≥µÏ†ú
bias_m_df = df_m_df.copy()

# Ìé∏Ìñ• Ïó¥ Ï∂îÍ∞Ä
bias_m_df['bias'] = 1
bias_m_df
```

```python
# ÌÉÄÍ≤ü Ïª¨Îüº ÏúÑÏπò Î≥ÄÍ≤ΩÏùÑ ÏúÑÌï¥ ÏÉàÎ°úÏö¥ ÌÉÄÍ≤ü Ïª¨Îüº Ï∂îÍ∞Ä ÌõÑ Í∏∞Ï°¥ ÌÉÄÍ≤ü Ï†úÍ±∞
bias_m_df['target'] = bias_m_df.Concretecompressivestrength
bias_m_df = bias_m_df.drop(labels='Concretecompressivestrength', axis=1)

bias_m_df
```

```python
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold

# Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ Î∂ÑÎ¶¨
# ÌîºÏ≤ò, ÌÉÄÍ≤ü Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨
features, targets = bias_m_df.iloc[:, :-1], bias_m_df.iloc[:, -1]

# Î¨∏Ï†ú/Ï†ïÎãµ Î∞è ÌïôÏäµ/ÌõàÎ†® Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨
X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

# Í≤ÄÏ¶ù Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨
val_X_train, val_X_test, val_y_train, val_y_test = \
train_test_split(X_train, y_train, test_size=0.2, random_state=124)

# ÌöåÍ∑Ä Î™®Îç∏ Îã¥Í∏∞
# reg_lambda: L2 Í∑úÏ†ú
k_xgb_r = XGBRegressor(random_state=124, reg_lambda=0.5)

# ÌååÎùºÎØ∏ÌÑ∞ Í∞í Ï°∞Ï†ï
# max_depth: ÏµúÎåÄ Ìä∏Î¶¨ ÍπäÏù¥ ÏÑ§Ï†ï, Ìä∏Î¶¨ Í∞íÏùÑ Ï§ÑÏó¨ÏÑú ÍπäÏù¥ Ï†úÌïú
# min_child_weight: ÏµúÏÜå Í∞ÄÏ§ëÏπò Ìï© ÏßÄÏ†ï, Í∞íÏùÑ ÎÜíÏó¨ ÏÑ∏Î∂ÄÏ†ÅÏúºÎ°ú ÎÇòÎâòÎäî Í≤ÉÏùÑ Î∞©ÏßÄÌïòÍ≥† Í≥ºÏ†ÅÌï©ÏùÑ Î∞©ÏßÄÌï®
parameters = {'max_depth': [1, 3], 'min_child_weight': [1, 3, 5], 'learning_rate': [0.15], 'n_estimators': [150, 200, 250]}

# ÍµêÏ∞®Í≤ÄÏ¶ù
# n_splits: Îç∞Ïù¥ÌÑ∞Î•º Î™á Í∞úÏùò Ìè¥ÎìúÎ°ú ÎÇòÎàåÏßÄÎ•º Í≤∞Ï†ï (ÏùºÎ∞òÏ†ÅÏúºÎ°ú 5 ÎòêÎäî 10)
# shuffle: Î∂ÑÌï† Ï†Ñ Îç∞Ïù¥ÌÑ∞ ÌòºÌï© Ïó¨Î∂Ä 
kfold = KFold(n_splits=10, random_state=124, shuffle=True)

# ÌïôÏäµ Î∞è ÍµêÏ∞® Í≤ÄÏ¶ù Î™®Îç∏ ÏÑ§Ï†ï
grid_xgb_r = GridSearchCV(k_xgb_r, param_grid=parameters, cv=kfold, n_jobs=-1)

# ÌõàÎ†®
grid_xgb_r.fit(X_train, y_train)
```
