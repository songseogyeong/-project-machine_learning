### üåü 1Cycle

#### üìä Îç∞Ïù¥ÌÑ∞ Ï†ÑÏ≤òÎ¶¨

```python
# Ï§ëÎ≥µÌñâ Ï†úÍ±∞ 
pre_c_df = c_df.drop_duplicates().reset_index(drop=True)
```

```python
# Ïª¨ÎüºÎ™Ö Î≥ÄÍ≤Ω
pre_c_df.columns = ['Cement', 'BlastFurnaceSlag', 'FlyAsh', 'Water', 'Superplasticizer',
                    'CoarseAggregate', 'FineAggregate', 'Age', 'Concretecompressivestrength']
```

```python
# ÌÉÄÍ≤ü Îç∞Ïù¥ÌÑ∞ Ïù¥ÏÉÅÏπò Ï†úÍ±∞
from sklearn.preprocessing import StandardScaler

# ÌëúÏ§ÄÌôî Í∞ùÏ≤¥ ÏÉùÏÑ±
std = StandardScaler()
# Îç∞Ïù¥ÌÑ∞ ÌëúÏ§ÄÌôî Î≥ÄÌôò
result = std.fit_transform(pre_c_df[['Concretecompressivestrength']])
# ÌëúÏ§ÄÌôî Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑ ÏÉùÏÑ±
std_c_df = pd.DataFrame(result, columns=['Concretecompressivestrength'])

# ÌÉÄÍ≤ü Îç∞Ïù¥ÌÑ∞ Ïù¥ÏÉÅÏπò Ï†úÍ±∞
std_c_df = std_c_df[std_c_df.Concretecompressivestrength.between(-1.96, 1.96)]

# Ïù¥ÏÉÅÏπò Ï†úÍ±∞ÌõÑ Ïú†ÏßÄÎêòÎäî Ïù∏Îç±Ïä§Î•º ÎåÄÏ°∞ÌïòÏó¨ ÏõêÎ≥∏ Îç∞Ïù¥ÌÑ∞ Ïù¥ÏÉÅÏπò Ï†úÍ±∞
pre_c_df = pre_c_df.iloc[std_c_df.index].reset_index(drop=True)
```

### üåü 3Cycle
K-Fold, GridSearchCVÎ•º ÌÜµÌïú ÍµêÏ∞®Í≤ÄÏ¶ù

```python
import numpy as np
from sklearn.metrics import mean_squared_error, mean_squared_log_error, r2_score

def get_evaluation(y_test, prediction):
    MSE = mean_squared_error(y_test, prediction)
    RMSE = np.sqrt(MSE)
    MSLE = mean_squared_log_error(y_test, prediction)
    RMSLE = np.sqrt(MSLE)
    R2 = r2_score(y_test, prediction)
    print('MSE: {:.4f}, RMSE: {:.4f}, MSLE: {:.4f}, RMSLE: {:.4f}, R2: {:.4f}'\
         .format(MSE, RMSE, MSLE, RMSLE, R2))
```

#### üìä KFold

```python
from sklearn.model_selection import train_test_split
from lightgbm import LGBMRegressor
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV

# Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ Î∂ÑÎ¶¨
# ÌîºÏ≤ò, ÌÉÄÍ≤ü Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨
features, targets = pre_c_df.iloc[:, :-1], pre_c_df.iloc[:, -1]

# Î¨∏Ï†ú/Ï†ïÎãµ Î∞è ÌïôÏäµ/ÌõàÎ†® Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨
X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

# Í∞ÄÏû• ÏÑ±Îä•Ïù¥ Ïö∞ÏàòÌïú Î™®Îç∏ Îã¥Í∏∞
lgb_r = LGBMRegressor(random_state=124)

# ÌååÎùºÎØ∏ÌÑ∞ Í∞í Ï°∞Ï†ï
parameters = {'n_estimators': [250, 300],'learning_rate': [0.2], 'max_depth': [20, 30]}

# ÍµêÏ∞®Í≤ÄÏ¶ù
kfold = KFold(n_splits=5, random_state=124, shuffle=True)

# ÌïôÏäµ Î∞è ÍµêÏ∞® Í≤ÄÏ¶ù Î™®Îç∏ ÏÑ§Ï†ï
grid_lgb_r = GridSearchCV(lgb_r, param_grid=parameters, cv=kfold, n_jobs=-1)

# ÌõàÎ†®
grid_lgb_r.fit(X_train, y_train)
```

```python
# ÌõàÎ†® Í≤∞Í≥º ÌôïÏù∏
result_df = pd.DataFrame(grid_lgb_r.cv_results_)[['params', 'mean_test_score', 'rank_test_score']]
result_df
```

```python
# ÏµúÏ†ÅÏùò Î™®Îç∏ Îã¥Í∏∞
grid_lgb_r = grid_lgb_r.best_estimator_

# ÏòàÏ∏°
prediction = grid_lgb_r.predict(X_test)

# ÌèâÍ∞Ä
get_evaluation(y_test, prediction)
```
